Para criar uma pequena estrutura para extrair os dados continuamente ao vivo.

```python
import time

time.time()
```

```python
# Se for necessário que se espere 15 minutos, o tempo final será a hora atual + 60 vezes 15 min.

t_end = time.time() + 60*1
t_end
```

```python
# Quote = cotação

df = pd.DataFrame(columns = ['DateTime', 'Quote'])
a = 0

while time.time() < t_end:
    
    x = si.get_live_price("PETR4.SA")
    
    d = {'DateTime': a, 'Quote': round(x,2)}
    # print(d)
    linha = pd.DataFrame(data = d, index = [a])
    
    a += 1
    
    df = pd.concat([df, linha], ignore_index = True)
    
    time.sleep(1)
```

```python
df.head()
```

```python
df.Quote.plot();
```

```python
from datetime import datetime
```

```python
agora = datetime.now()
```

```python
agora.strftime("%d-%b-%Y %H:%M:%S")
```

```python
t_end = time.time() + 60*1

df = pd.DataFrame(columns = ['DateTime', 'Quote'])
a = 0

while time.time() < t_end:
    
    x = si.get_live_price("PETR4.SA")
    dateTimeObj = datetime.now()
    timestampStr = dateTimeObj.strftime("%d-%b-%Y %H:%M:%S")
    
    d = {'DateTime': timestampStr, 'Quote': round(x,2)}
    
    linha = pd.DataFrame(data = d, index = [a])
    
    a += 1
    
    df = pd.concat([df, linha], ignore_index = True)
    
    time.sleep(1)
```

```python
df.head()
```

```python
df.index = df.DateTime
```

```python
df.Quote.plot(figsize = (12,8));
```
